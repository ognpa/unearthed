# Read all data and filter when startvelocity end velocity is 0 and riser height 0

# Flag overweight truck 

```python
df['theta']=(df.RiseHeight /df.SlopeLength).apply(lambda x:np.arcsin(x))
df['OW']=0
df.loc[(df.Payload+df['Empty Weight (kg)'])>=df['Max Gross OperatingWeight'],'OW']=1
```

```python
tgt_set=df.loc[:,['OW','Truck','theta','RiseHeight','SlopeLength','Payload','StartVelocity','EndVelocity','DurationCalculated','TargetDuration','X1','Y1','Z1','X2','Y2','Z2']]
```

```python
d=tgt_set.loc[:,['OW','theta','Truck','RiseHeight','SlopeLength','TargetDuration',
                 'StartVelocity','EndVelocity','Tonnage per hour',
                 'X1','Y1','Z1','X2','Y2','Z2']]



tc_df=pd.read_csv("Truck Cycle.csv")
tc_df.head()
tc_trip_count=tc_df.groupby("Truck",as_index=False).Date.count()
tc_trip_count.columns=['Truck','Truck_trip_count']
d=d.merge(tc_trip_count)
```

# Models-more to be added and tuned

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor

from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV
estimators = [

    {'name': 'RandomForest',

'instance': RandomForestRegressor(n_estimators=300, oob_score=True, n_jobs = -1, random_state=42,

             criterion= "mse", max_features = "auto", min_samples_leaf = 1)},

    {'name': 'GBR',

'instance': GradientBoostingRegressor(loss='ls', alpha=0.95, n_estimators=300)},

    {'name': 'ABR',

'instance': AdaBoostRegressor(DecisionTreeRegressor(max_depth=4),

n_estimators=300, random_state=np.random.RandomState(1))},

    {'name': 'LSR',

'instance': Lasso(alpha=1e-8,normalize=True, max_iter=1e5)},

    {'name': 'ETR',

'instance': ExtraTreesRegressor(n_estimators=300)}

    ]


def get_best_params(X, y):

    param_grid = { "n_estimators" : [200, 300, 500,600],

    "max_depth" : [3, 5,7],

    "oob_score" : [True, False]

    }

    rf = RandomForestRegressor(n_estimators=200, oob_score=True, n_jobs = -1, random_state=42, bootstrap='True',

    criterion= "mse", max_features = "auto", min_samples_leaf = 50)

    CV_rfc = GridSearchCV(estimator=rf, param_grid=param_grid, cv= 10)

    CV_rfc.fit(X, y)

    print (CV_rfc.best_params_)

    return CV_rfc.best_params_


est = {}
alldf={}
modeldf={}
def train_data(X_train, y_train, X_val, y_val, estimators):


    for estimator in estimators:
        print(estimator['name'])

        model1=estimator['instance'].fit(X_train, y_train)

        cv = RepeatedStratifiedKFold(n_splits=2, n_repeats=10, random_state=42)

        val_errs = estimator['instance'].score(X_val,y_val)

        print(f"validation error: {val_errs.mean()}, std dev: {val_errs.std()}")


        est[estimator['instance']] = val_errs.mean()
        alldf[estimator['name']]=val_errs
        modeldf[estimator['name']]=model1
        model = min(iter(est.keys()), key=lambda k: est[k])

        print("rsquared", est[model])




train_data(X_train,y_train.values,X_test,y_test.values,estimators)
```

# Test set for every 1m increase in riser height what is the tonnage per hour


```python
#As we go on increasing riser height by 1 m for every truck we want to predict tonnageperhour given target duration is mean for that truck


mean_tgt_duration=tgt_set.groupby('Truck',as_index=False).TargetDuration.mean()
mean_tgt_duration.head()
```

```python
owcount=tgt_set.groupby('Truck',as_index=False)['OW'].count()
owcount.head()
```

```python
reqd=tgt_set.groupby('Truck',as_index=False).RiseHeight.max()
reqd['RiseHeight']=reqd['RiseHeight']+1
reqd
```

```python
total_df=mean_tgt_duration.merge(reqd,on='Truck').merge(owcount)
```

```python
#Mean gradient by truck
gradient=tgt_set.groupby('Truck',as_index=False).theta.mean()
gradient
gradient.theta=1/gradient.theta
gradient.columns=['Truck','Total_increase_per_1_m']
total_df=total_df.merge(gradient,on='Truck')
```

```python
slope_mean=tgt_set.groupby('Truck',as_index=False).SlopeLength.mean()
total_df=total_df.merge(slope_mean,on='Truck')
total_df['NewSlopeLength']=total_df.SlopeLength+total_df.Total_increase_per_1_m

total_df.drop('SlopeLength',axis=1,inplace=True)
total_df
```

```python
total_df.rename(columns={"NewSlopeLength":"SlopeLength"},inplace=True)
total_df.drop('Total_increase_per_1_m',axis=1,inplace=True)
```

```python
start_velocity=tgt_set.groupby('Truck',as_index=False).StartVelocity.mean()
end_velocity=tgt_set.groupby('Truck',as_index=False).EndVelocity.mean()

total_df=total_df.merge(start_velocity,on='Truck').merge(end_velocity,on='Truck')
total_df.head()
```

```python
lastX1=tgt_set.groupby('Truck',as_index=False).X1.last()
lastY1=tgt_set.groupby('Truck',as_index=False).Y1.last()
lastZ1=tgt_set.groupby('Truck',as_index=False).Z1.last()
lastX2=tgt_set.groupby('Truck',as_index=False).X2.last()
lastY2=tgt_set.groupby('Truck',as_index=False).Y2.last()
lastZ2=tgt_set.groupby('Truck',as_index=False).Z2.last()
coords=lastX1.merge(lastY1).merge(lastZ1).merge(lastX2).merge(lastY2).merge(lastZ2)
total_df=total_df.merge(coords)

```

```python
total_df=total_df.merge(tc_trip_count)
total_df
```

```python
total_df['Truck']=total_df.Truck.astype('category')
total_df=pd.get_dummies(total_df)
total_df.head()
```
