```python
from scipy.interpolate import spline
import numpy as np
import pandas as pd
df=pd.read_excel("Road Segment Data.xlsx")
df=df.loc[df.RiseHeight>0]
truck_spec=pd.read_excel("Truck_Spec.xlsx",sheet="Truck Spec")
truck_spec['Truck ID']=truck_spec['Truck ID'].apply(lambda x:'TRH'+str(x))
df=df.merge(truck_spec,left_on='Truck',right_on='Truck ID')
df=df.loc[((df.StartVelocity>0)&(df.EndVelocity>0))]
df.head()
```

```python
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
matplotlib.__version__
```

```python
def threedplot(x1,y1,z1,x2,y2,z2,truck):
    threedee = plt.figure().gca(projection='3d')
    threedee.plot(x1,y1,z1,x2,y2,z2)
    
    threedee.set_xlabel('x')
    threedee.set_ylabel('y')
    threedee.set_zlabel('z')
    threedee.set_title(truck)
    plt.show()
    
import matplotlib.pyplot as plt
df.groupby('Truck').apply(lambda x:threedplot(x.X1,x.Y1,x.Z1,x.X2,x.Y2,x.Z2,x.Truck.unique()))
```

```python
df['theta']=(df.RiseHeight /df.SlopeLength).apply(lambda x:np.arcsin(x))
df['OW']=0
df.loc[(df.Payload+df['Empty Weight (kg)'])>=df['Max Gross OperatingWeight'],'OW']=1
```

```python
tgt_set=df.loc[:,['OW','Truck','theta','RiseHeight','SlopeLength','Payload','StartVelocity','EndVelocity','DurationCalculated','TargetDuration','X1','Y1','Z1','X2','Y2','Z2']]
```

```python
tgt_set['Tonnage per hour']=tgt_set.Payload*60/tgt_set.DurationCalculated
tgt_set.head()
```

```python
d=tgt_set.loc[:,['OW','Truck','theta','RiseHeight','SlopeLength','TargetDuration',
                 'StartVelocity','EndVelocity','Tonnage per hour',
                 'X1','Y1','Z1','X2','Y2','Z2']]



```

```python
d['Truck']=d.Truck.astype('category')
d=pd.get_dummies(d)
```

```python
reqd=[i for i in d.columns if (i!='Tonnage per hour')]
reqd
```

```python
from sklearn.model_selection import train_test_split
X=d.loc[:,reqd]
y=d['Tonnage per hour']
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=32,shuffle=True)
X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_auc_score
rf=RandomForestRegressor(n_estimators=200)
rf.fit(X_train,y_train)
```

```python
sorted(zip(map(lambda x: round(x, 4), rf.feature_importances_*100), reqd), reverse=True)
```

```python
preds=rf.predict(X_test)
df=pd.DataFrame(preds)
df.columns=['pred']
rf.score(X_test,y_test)
```
