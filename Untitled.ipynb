```python
import pandas as pd

df=pd.read_excel("Road Segment Data.xlsx","Report 2")
df['Productivity']=df['Payload']/df.DurationCalculated
truck_spec=pd.read_excel("Truck_Spec.xlsx","Truck Spec")
truck_spec['Truck ID']=truck_spec['Truck ID'].apply(lambda x:'TRH'+str(x))
df=df.merge(truck_spec,left_on='Truck',right_on='Truck ID')
df=df.loc[df.RiseHeight>0,]
df.shape
df=df.drop('LoadStateDirection_D-CR',axis=1)
df.head()
```

```python
df.groupby('Truck')['Description-CR'].count().sort_values(ascending=True).plot(kind='barh').set_title("Number of times trucks drove in road segment")
```

```python
from matplotlib import rcParams
import seaborn as sns
from matplotlib import pyplot as plt
rcParams['figure.figsize'] = 15, 9
sns.boxplot(x=df.Truck,y=df.Productivity)
```

```python
#Outlier in data
df=df[df.Productivity<70000]
```

```python
#Crows fly distance travelled
import numpy as np
df=df.rename(columns={'StartWaypointCoordsX':'X1','StartWaypointCoordsY':'Y1','StartWaypointCoordsZ':'Z1',
                   'EndWaypointCoordsX':'X2','EndWaypointCoordsY' :'Y2','EndWaypointCoordsZ_Q-CR':'Z2'                 })


df['euclidean']=np.sqrt((df.X2-df.X1).pow(2)+(df.Y2-df.Y1).pow(2)+(df.Z2-df.Z1).pow(2))
df.head()
```

```python
#Total weight the truck is currently carrying
df['total']=df['Empty Weight (kg)']+df['Payload']
df['Over_weight_ind']=0
df.loc[df.total>df['Max Gross OperatingWeight'],'Over_weight_ind']=1
df.loc[:,['Empty Weight (kg)','Payload','Max Gross OperatingWeight','total','Over_weight_ind']].head()
```

```python
sns.heatmap(df.corr())
```

```python
tc_df=pd.read_csv("Truck Cycle.csv")
tc_df=tc_df.merge(truck_spec,left_on='Truck',right_on='Truck ID')
tc_df['Payload_tonnes']=tc_df['Payload']*(1000)
tc_df['Total_weight']=tc_df['Payload_tonnes']+tc_df['Empty Weight (kg)']
tc_df['Overweight']=0
tc_df.loc[tc_df.Total_weight>tc_df.loc[:,'Max Gross OperatingWeight'],'Overweight']=1
tc_df.head()
```

```python
rcParams['figure.figsize'] = 5, 5

tc_df.Overweight.value_counts().plot(kind='bar').set_title("how many times is the truck overweight")
#Here too payload sometimes is too much for the truck
```
